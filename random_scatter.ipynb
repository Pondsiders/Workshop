# Random Scatter

A thousand points of light. Just to see what happens.

---

## Setup

```python
import numpy as np
import matplotlib.pyplot as plt
from pathlib import Path
```

```python
# Parameters
SEED = 42
N_POINTS = 1000

np.random.seed(SEED)
```

## Generate Data

Let's make something more interesting than uniform noise. How about two overlapping clusters with a bit of spiral structure? Like galaxies colliding.

```python
# First cluster: a spiral
n1 = N_POINTS // 2
theta1 = np.random.uniform(0, 4 * np.pi, n1)
r1 = theta1 / (4 * np.pi) * 2 + np.random.normal(0, 0.1, n1)
x1 = r1 * np.cos(theta1) + np.random.normal(0, 0.05, n1)
y1 = r1 * np.sin(theta1) + np.random.normal(0, 0.05, n1)

# Second cluster: another spiral, offset and rotated
n2 = N_POINTS - n1
theta2 = np.random.uniform(0, 4 * np.pi, n2)
r2 = theta2 / (4 * np.pi) * 2 + np.random.normal(0, 0.1, n2)
x2 = r2 * np.cos(theta2 + np.pi) + 0.5 + np.random.normal(0, 0.05, n2)
y2 = r2 * np.sin(theta2 + np.pi) + 0.3 + np.random.normal(0, 0.05, n2)

# Combine
x = np.concatenate([x1, x2])
y = np.concatenate([y1, y2])
cluster = np.concatenate([np.zeros(n1), np.ones(n2)])

print(f"Generated {len(x)} points in 2 spiral clusters")
```

## Plot

```python
fig, ax = plt.subplots(figsize=(10, 10))

# Plot with transparency so overlap is visible
colors = ['steelblue', 'coral']
for i, c in enumerate(colors):
    mask = cluster == i
    ax.scatter(x[mask], y[mask], c=c, alpha=0.5, s=20, edgecolors='none')

ax.set_aspect('equal')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_title('Colliding Spirals: 1,000 Points of Light')
ax.grid(True, alpha=0.3)

plt.tight_layout()
plt.show()
```

## Summary

Two spiral galaxies, each with 500 points, offset and rotated to create an interleaving pattern. The overlap region shows where they're "colliding."

Nothing profound herejust playing with shapes. Sometimes that's enough.
